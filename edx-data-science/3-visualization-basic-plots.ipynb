# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot') # 
```

## Histograms

### Students data

```python
student_dataset = pd.read_csv('./data/students.data', index_col=0)
student_dataset.head()
```

```python
my_series = student_dataset.G3  # series 
my_series.plot.hist(alpha=0.5)
plt.xlabel('G3')
plt.ylabel('# of students')
plt.title('Student G3')
#my_series.plot.hist(alpha=0.5,bins=20)
#my_series.plot.hist(alpha=0.5,bins=[0,5,10,15,20])

```

```python
my_dataframe = student_dataset[['G3','G2','G1']] # extract columns into new data frame 
#my_dataframe.plot.hist(alpha=0.5)
#my_dataframe.plot.hist(alpha=0.5, normed=True)
#my_dataframe.plot.hist(alpha=0.5, normed=True, cumulative=False)
#my_dataframe.plot.hist(alpha=0.5, stacked=True)
#my_dataframe.plot.hist(alpha=0.5, stacked=True, log=True)
#my_dataframe.plot.hist(alpha=0.5, stacked=True, orientation='horizontal')
my_dataframe.plot.hist(alpha=0.9, stacked=False, orientation='horizontal', histtype='step')
```

### Wheat data

```python
df = pd.read_csv('./data/wheat.data', index_col=0)
df.head()
```

```python
df[df.wheat_type=='rosa'].asymmetry.plot.hist(alpha=0.4)
df[df.wheat_type=='kama'].asymmetry.plot.hist(alpha=0.4)
df[df.wheat_type=='canadian'].asymmetry.plot.hist(alpha=0.4)
plt.legend(['rosa','kama','canadian'])
```
