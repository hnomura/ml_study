# Higher Dimensionality Visualizations

## Pandas Parallel Coordinates

```python
from sklearn.datasets import load_iris
#from pandas.tools.plotting import parallel_coordinates  - will be deprecated
from pandas.plotting import parallel_coordinates

import pandas as pd
import matplotlib.pyplot as plt 

plt.style.use('ggplot')
```

```python
# Scikit-Learn IRIS data
data = load_iris()
dir(data)
```

```python
# Target name is categorical text 
data.target_names
```

```python
# Target number can be used to index into target names
data.target
```

```python
# load data into data frame
df = pd.DataFrame(data.data, columns=data.feature_names)

# create new column to store target name 
df['target_names'] = [data.target_names[i] for i in data.target]
```

```python
plt.figure()
parallel_coordinates(df, 'target_names')  # grouped by target_names
```

## Andrews Plot (Curves)

```python
# Andrews Curve : Feature values as if they were Fourier Series Coefficients
#                  A technique to visualize distinction of different clusters...
#
# for X(i,:) [ith sample], plot is the following f(t) in [-pi, +pi] region
# f(t)=X(i,1)/G2+X(i,2)sin(2t)+X(i,3)cos(2t)+

from pandas.plotting import andrews_curves
plt.figure()
andrews_curves(df, 'target_names')
```

## Scatter Matrix  
Diagonal part of the plots can be either KDE (Kernel Density Esimation) or plain Histogram 

```python
from pandas.plotting import scatter_matrix

# diagonal='kde' : Kernel Density Estimation
# diagonal='hist': Histogram (bar graph)

axes= scatter_matrix(df, figsize=(12, 12), diagonal='kde')
```

## imshow 

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(1000,5), columns=['a','b','c','d','e'])

# pair-wise correlation between different features
df.corr()
```

```python
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

# interpolation is irrelevant here 
# interpolation = ['none', 'nearest', ... ]
# see https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html#matplotlib.pyplot.imshow
plt.imshow(df.corr(), cmap=plt.cm.RdBu_r, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)
```
