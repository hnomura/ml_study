# L1 regularization  
With X(NxD), only a part of D features contribute to result. Sparsity. 

L1 regularlization suites for sparsity case, i.e., many of weight will be zero.   
L2 regularlization suites for weight being very small, assumption used to overfitting to outliers  
Combination of L1 and L2 is called ElasticNet


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Synthetic Data
# FAT matrix
N = 50
D = 50

# Input
X = (np.random.random((N,D)) - 0.5)*10

# True W 
true_w = np.array([1, 0.5, -0.5] + [0]*(D-3))

# Target synthesized with Truw W
Y = X.dot(true_w) + np.random.randn(N)*0.5
```

```python
# Gradient Descent
costs = []
w = np.random.randn(D)/np.sqrt(D)
learning_rate = 0.001
l1 = 10.0

for t in range(500):
    Yhat = X.dot(w)
    delta = Yhat - Y
    w = w - learning_rate * (X.T.dot(delta) + l1*np.sign(w))
    
    mse = delta.dot(delta) / N 
    costs.append(mse)
    
print('final w = ', w)
```

```python
plt.plot(costs)
plt.show()
```

```python
plt.plot(true_w, label='true w')
plt.plot(w, label='w_map')
plt.legend()
plt.show()
```
