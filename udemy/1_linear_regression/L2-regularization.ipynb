# L2 Regularization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
X = np.linspace(0,10,N)
Y = 0.5 * X + np.random.randn(N)
```

```python
# Outliers !
Y[-1] += 30
Y[-2] += 30
```

```python
plt.scatter(X,Y)
plt.title('original input data')
plt.show()
```

```python
# add bias term to X 
X2 = np.vstack([np.ones(N), X]).T
X2
```

```python
# Without L2 regularization
w_ml = np.linalg.solve(X2.T.dot(X2), X2.T.dot(Y))
Yhat_ml = X2.dot(w_ml)
```

```python
plt.scatter(X2[:,1],Y)
plt.title('normal MLR with maximum likelihood')
plt.plot(X2[:,1],Yhat_ml, color='g')
plt.show()
```

```python
# With L2 regularization (MAP = MAximizing Posterium...? ) : Ridge Regression
l2 = 1000.0
w_map = np.linalg.solve(l2*np.eye(2) + X2.T.dot(X2), X2.T.dot(Y))
Yhat_map = X2.dot(w_map)
```

```python
plt.scatter(X2[:,1], Y)
plt.plot(X2[:,1], Yhat_ml, color='g', label='maximum likelihood')
plt.plot(X2[:,1], Yhat_map, color='r', label='ridge regression with l2=%d' %l2)
plt.legend()
plt.show()
```
